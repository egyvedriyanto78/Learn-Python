## Import Modules

## Google Books API

### Request

```python
# Meminta request informasi buku dengan ID 'Pz4YDQAAQBAJ'
response = requests.get('https://www.googleapis.com/books/v1/volumes/Pz4YDQAAQBAJ?key=' + au.gbooks_key)
```

### Print Result

```python
# Mengembalikan respon yang didapat dengan JSON
response_json = response.json()

# Agar hasil print rapi saja
pret = json.dumps(response_json, indent=4)
print(pret)
```

### Request

```python
# Meminta request informasi buku dengan Search Terms (query) 'harry potter'
response2 = requests.get('https://www.googleapis.com/books/v1/volumes?q=harry+potter&key=' + au.gbooks_key)
```

### Print Result

```python
# Mengembalikan respon yang didapat dengan JSON
response_json_2 = response2.json()

# Agar hasil print rapi saja
pret2 = json.dumps(response_json_2, indent=4)
print(pret2)
```

## Twitter API

### Import Modules

```python
# Import Python library for accessing Twitter API
import tweepy

# Import module yang berisi credentials Twitter API (access token, etc), module buatan sendiri
import authentication as au
```

### Autentikasi

```python
# Setting credentials agar dapat digunakan untuk mengakses Twitter
auth = tweepy.OAuthHandler(au.consumer_key, au.consumer_secret)
auth.set_access_token(au.access_token, au.access_secret)

api = tweepy.API(auth)
```

### Exploring the API

#### Home Timeline

```python
# Mengambil tweets dari timeline akun yang terhubung dengan API credentials tadi
public_tweets = api.home_timeline()
for tweet in public_tweets:
  print(tweet.text)
```

#### Get User

```python
# Cari user yang memiliki screen name 'foodfess2'
user = api.get_user(screen_name='foodfess2')
```

#### Show User's Friends

```python
print('Screen name:', user.screen_name)
print('Followers:', user.followers_count)

# Print screen name dari Friends 'foodfess2'
for friend in user.friends():
  print(friend.screen_name)
```

#### Twitter Trending Topic

```python
# Melihat trending topic Twitter beserta lokasinya
api.trends_available()
```

#### Twitter Trending Topic Based on Location

```python
# Melihat trending topic Twitter dari lokasi terdekat spesifik
lat = -0.789275
long = 113.921327

api.trends_closest(lat, long)
```

#### Twitter Trending Topic Based on Place

```python
# Melihat 50 trending topic teratas dari lokasi spesifik berdasarkan WOEID
api.trends_place(23424846)
```
